# now time for visualzation to get insights 

##  first time series plots to see global vs local changes over time :

```python
# Convert the 'Date' column to a datetime object
df['Date'] = pd.to_datetime(df['Date'])

# Set the 'Date' column as the index
df.set_index('Date', inplace=True)
```

```python
# Create subplots for local and global prices
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))
fig.suptitle('Gold Prices Over Time')

# Separate DataFrames for local and global prices
local_df = df[[col for col in df.columns if 'Local Price' in col]]
global_df = df[[col for col in df.columns if 'Global Price' in col]]

# Plot local gold prices
local_df.plot(ax=axes[0], label=local_df.columns)
axes[0].set_title('Local Prices')
axes[0].set_ylabel('Price')

# Plot global gold prices
global_df.plot(ax=axes[1], label=global_df.columns)
axes[1].set_title('Global Prices')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Price')

# Add legends
axes[0].legend()
axes[1].legend()

# Adjust spacing between subplots
plt.tight_layout()
plt.subplots_adjust(top=0.9)

# Show the plot
plt.show()
```

#### we can see that local got alot of ups prices kind of shited massively while global got only one big change in 1-2023


```python
# Create boxplots for local and global prices
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 8))
fig.suptitle('Boxplots of Gold Prices (Local and Global)')
local_buy_df= local_df[[col for col in local_df.columns if 'Local Price/Buy' in col]]
local_sell_df= local_df[[col for col in local_df.columns if 'Local Price/sell' in col]]

# Plot boxplot for local gold prices
local_buy_df.boxplot(ax=axes[0], grid=False)
axes[0].set_title('Local Prices/buy')
axes[0].set_ylabel('Price')

local_buy_df.boxplot(ax=axes[1], grid=False)
axes[1].set_title('Local Prices/sell')
axes[1].set_ylabel('Price')
# Plot boxplot for global gold prices
global_df.boxplot(ax=axes[2], grid=False)
axes[2].set_title('Global Prices')
axes[2].set_ylabel('Price')

# Adjust spacing between subplots
plt.tight_layout()
plt.subplots_adjust(top=0.9)

# Show the plot
plt.show()
 
```
